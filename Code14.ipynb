```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Read the data
with open('Week14Assignment.txt', 'r') as file:
    data = file.readlines()[1:]  # Skip the first line

# Initialize lists for each category
readmitted = []
staff_satisfaction = []
cleanliness_satisfaction = []
food_satisfaction = []
comfort_satisfaction = []
communication_satisfaction = []

# Parse the data
for line in data:
    patient_data = line.split(',')
    readmitted.append(int(patient_data[1]))
    staff_satisfaction.append(float(patient_data[2]))
    cleanliness_satisfaction.append(float(patient_data[3]))
    food_satisfaction.append(float(patient_data[4]))
    comfort_satisfaction.append(float(patient_data[5]))
    communication_satisfaction.append(float(patient_data[6]))

# Calculate and display the number of patients who were readmitted and the average satisfaction score for each category
print("Number of patients who were readmitted:", sum(readmitted))
print("Average staff satisfaction score:", np.mean(staff_satisfaction))
print("Average cleanliness satisfaction score:", np.mean(cleanliness_satisfaction))
print("Average food satisfaction score:", np.mean(food_satisfaction))
print("Average comfort satisfaction score:", np.mean(comfort_satisfaction))
print("Average communication satisfaction score:", np.mean(communication_satisfaction))

# Implement logistic regression
X = np.array([staff_satisfaction, cleanliness_satisfaction, food_satisfaction, comfort_satisfaction, communication_satisfaction]).T
y = np.array(readmitted)
clf = LogisticRegression(random_state=0).fit(X, y)

# Display the logistic regression results
print("Logistic Regression Coefficients:", clf.coef_)
print("Logistic Regression Intercept:", clf.intercept_)

# Plot the data points along with the logistic regression curve
plt.scatter(X[:, 0], y)
plt.plot(X[:, 0], clf.predict_proba(X)[:, 1])
plt.show()
```
